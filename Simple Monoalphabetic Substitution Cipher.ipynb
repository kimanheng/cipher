# Simple Monoalphabetic Subsitution Cipher

## Configuration 

```python
config = {
    'a': 'q',
    'b': 'w',
    'c': 'e',
    'd': 'r',
    'e': 't',
    'f': 'y',
    'g': 'u',
    'h': 'i',
    'i': 'o',
    'j': 'p',
    'k': 'a',
    'l': 's',
    'm': 'd',
    'n': 'f',
    'o': 'g',
    'p': 'h',
    'q': 'j',
    'r': 'k',
    's': 'l',
    't': 'z',
    'u': 'x',
    'v': 'c',
    'w': 'v',
    'x': 'b',
    'y': 'n',
    'z': 'm'
}
```

## Encryption Function

```python
def encrypt(plain_msg: str) -> str: 
    encrypted_msg = ""
    for c in plain_msg: 
        if c.islower():
            encrypted_msg += config[c]
        elif c.isupper():
            encrypted_msg += config[c.lower()].upper()
        else:
            encrypted_msg += c
    return encrypted_msg

encrypt("In this enigmatic world, every encoded message becomes a ciphered sonnet, challenging minds to unravel its secrets while defenders strive to ensure that the cryptographic ballet remains an unyielding fortress for the safeguarding of information.")
```
