## Encryption Function

```python
def encrypt(message, key):
    encrypted = "" 
    for _ in message:  
        if _.isupper():
            encrypted += chr((ord(_) - ord('A') + key) % 26 + ord('A'))
        elif _.islower():
            encrypted += chr((ord(_) - ord('a') + key) % 26 + ord('a'))
        else:
            encrypted += _
    return encrypted

encrypt("In this enigmatic world, every encoded message becomes a ciphered sonnet, challenging minds to unravel its secrets while defenders strive to ensure that the cryptographic ballet remains an unyielding fortress for the safeguarding of information.", 5)
```

# Decryption Function

```python
def decrypt(message, key):
    decrypted = ""
    for _ in message:
        if _.isupper():
            decrypted += chr((ord(_) - ord('A') - key) % 26 + ord('A'))
        elif _.islower():
            decrypted += chr((ord(_) - ord('a') - key) % 26 + ord('a'))
        else:
            decrypted += _
    return decrypted

decrypt("Ns ymnx jsnlrfynh btwqi, jajwd jshtiji rjxxflj gjhtrjx f hnumjwji xtssjy, hmfqqjslnsl rnsix yt zswfajq nyx xjhwjyx bmnqj ijkjsijwx xywnaj yt jsxzwj ymfy ymj hwduytlwfumnh gfqqjy wjrfnsx fs zsdnjqinsl ktwywjxx ktw ymj xfkjlzfwinsl tk nsktwrfynts.", 5)
```
